```python

import ROOT as r
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
## Here i define all the parameters needed for the plot

T = 0.200 #arbitrairy temperature in Gev
Rap_count = 1000 #arbitrairy nb of count for 1 event in a given window of rapidity
p_mass = 0.938 #mass proton in Gev/c2, i will import it from root librairy
```

```python
## Definition of the exponential function as in the article

def exponential(pt, count, temperature, mass):
    value = count * pt * (1/(temperature*(mass+temperature))) * m.exp(-(m.sqrt((pt)**2+(mass)**2))/temperature)
    return value

## Definition of the Boltzmann function

def Boltzmann(pt, temperature, mass):
    value = 4100 * (m.sqrt((pt)**2+(mass)**2)) * m.exp(-(m.sqrt((pt)**2+(mass)**2))/temperature) ##the 4100 is an arbitrairy rescale value to compare behaviours
    return value
    
    
```

```python

test = exponential(10,1000,100,938)
test2 = Boltzmann(10,100,938)
print(test, test2)
```

```python
L = 1000

pt_list = np.linspace(0.3,20,L)
exp_list = np.zeros(L)
boltz_list = np.zeros(L)

for i in range(L):
    exp_list[i] = exponential(pt_list[i], Rap_count, T, p_mass)
    boltz_list[i] = Boltzmann(pt_list[i], T, p_mass)

#print(exp_list)
```

```python
plt.plot(pt_list, exp_list)
plt.plot(pt_list, boltz_list)

plt.grid()
#plt.axes()


```
